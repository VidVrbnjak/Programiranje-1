## Predpriprava

```python
import explode as explode
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
dirpath = os.getcwd()
pd.options.display.max_rows = 30
filename = dirpath + "\koncani_podatki.csv"
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
df = pd.read_csv('koncani_podatki.csv', skiprows=[0], delimiter=',',
                 names=['znamka', 'letnik', 'prevozenih', 'tip_motorja', 'moc_motorja', 'menjalnik', 'cena'])
```

Vidimo, da Audi, BMW in Mercedes-Benz predstavljajo več kot polovico vseh zajetih oglasov, kar je tudi pričakovano saj spadajo v višji cenovni razred kot naprimer Chevrolet, ki se pojavi le enkrat, so pa tudi zadosti pogosti pri nas. Ker naprimer Jaguar ni zelo pogost avto v Sloveniji spada pa tudi v višji cenovni razred se pojavi samo štirikrat.

```python
avaragePrice = df[['znamka', 'cena']].groupby(['znamka']).mean()
avaragePrice.rename(columns={"cena": "povprecna_cena"}, inplace=True)

avgBC = avaragePrice.plot(kind='bar', title="Povprečna cena glede na znamko",
                          figsize=(15, 10), legend=False, fontsize=12,   edgecolor='blue')
avgBC.set_xlabel("Znamka", fontsize=12)
avgBC.set_ylabel("Cena [ x1000€]", fontsize=12)

avgBC.spines['right'].set_visible(False)
avgBC.spines['top'].set_visible(False)
avgBC.tick_params(color='gray', labelcolor='gray')
for spine in avgBC.spines.values():
    spine.set_edgecolor('gray')
```

## Popvrečno število prevoženih kilometrov glede na letnik

```python
avarageMileage = df[['letnik', 'prevozenih']].groupby(['letnik']).mean()

avgMilBC = avarageMileage.plot(kind='bar', title="Povprečna kilometrina glede na letnik",
                               figsize=(15, 10), legend=False, fontsize=12,
                               color='#ff9999', edgecolor='red')
avgMilBC.set_xlabel("Letnik avta", fontsize=12)
avgMilBC.set_ylabel("Prevoženi km", fontsize=12)
```

Opazimo da letnik avtomobila nima prevelikega vpliva na število prevoženih kilometrov, razen za avte, ki so od letnika 2019 naprej.

```python
df.sort_values('letnik', ascending = True).plot(x='letnik', y='cena', kind='scatter')
```

## Porazdelitev tipov motorja

```python
tipMotorja = df
tipMotorja['stevilo'] = 0 
tipMotorja = df[['tip_motorja',
                 'stevilo']].groupby(['tip_motorja']).count().plot.pie(subplots=True,
                                                                       autopct='%.1f%%', colors=colors,
                                                                       pctdistance=0.85,
                                                                       explode = (0.05, 0.05, 0.05, 0.05))
plt.legend(loc='best', bbox_to_anchor=(0.5, 0., 0.5, 0.5))
plt.title("Tip motorja")
plt.ylabel('') 
```

Najpogostejši so avtomobili z dieselskim motorjem.

## Porazdelitev menjalnikov

```python
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#E7EFC5', '#BFD7B5']
tipMenjalnika = df  
tipMenjalnika['stevilo'] = 0 
tipMenjalnika = df[['menjalnik', 'stevilo']].groupby(['menjalnik']).count()
tipMenjalnika = tipMenjalnika.sample(frac=1)
tipMenjalnika.plot.pie(subplots=True, colors=colors, autopct='%.1f%%', fontsize=7, figsize=(10, 8),
                       explode=(0.03, 0.04, 0.03, 0.02, 0.01, 0.01), shadow=False)
plt.ylabel('') 
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.legend(loc='best', bbox_to_anchor=(0.7, 0.5, 0.5, 0.5))  
plt.title("Tip menjalnika")  
```

Zanimiva opazka, da so med najdražjimi avtomibli skoraj vsi z avtomatskim menjalnikom.
